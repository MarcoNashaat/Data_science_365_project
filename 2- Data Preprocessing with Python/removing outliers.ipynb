```python
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#loading datasets

df_1 = pd.read_csv("minutes_watched_2021_paid_0.csv")
df_2 = pd.read_csv("minutes_watched_2021_paid_1.csv")
df_3 = pd.read_csv("minutes_watched_2022_paid_0.csv")
df_4 = pd.read_csv("minutes_watched_2022_paid_1.csv")
```

```python
#plotting minutes_watched column distribution

# Create a figure and axes for subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Plot KDE for each DataFrame in separate subplots
sb.kdeplot(df_1['minutes_watched'], ax=axes[0, 0])
axes[0, 0].set_title('minutes_watched_2021_paid_0')

sb.kdeplot(df_2['minutes_watched'], ax=axes[0, 1])
axes[0, 1].set_title('minutes_watched_2021_paid_1')

sb.kdeplot(df_3['minutes_watched'], ax=axes[1, 0])
axes[1, 0].set_title('minutes_watched_2022_paid_0')

sb.kdeplot(df_4['minutes_watched'], ax=axes[1, 1])
axes[1, 1].set_title('minutes_watched_2022_paid_1')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
#remove outliers

# Calculate the 99th percentile value and filter out outliers for each DataFrame

# For df_1
percentile_99_df1 = df_1['minutes_watched'].quantile(0.99)
df_1_filtered = df_1[df_1['minutes_watched'] < percentile_99_df1]

# For df_2
percentile_99_df2 = df_2['minutes_watched'].quantile(0.99)
df_2_filtered = df_2[df_2['minutes_watched'] < percentile_99_df2]

# For df_3
percentile_99_df3 = df_3['minutes_watched'].quantile(0.99)
df_3_filtered = df_3[df_3['minutes_watched'] < percentile_99_df3]

# For df_4
percentile_99_df4 = df_4['minutes_watched'].quantile(0.99)
df_4_filtered = df_4[df_4['minutes_watched'] < percentile_99_df4]
```

```python
#plotting minutes_watched column distribution after removing outliers

# Create a figure and axes for subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Plot KDE for each DataFrame in separate subplots
sb.kdeplot(df_1_filtered['minutes_watched'], ax=axes[0, 0])
axes[0, 0].set_title('minutes_watched_2021_paid_0')

sb.kdeplot(df_2_filtered['minutes_watched'], ax=axes[0, 1])
axes[0, 1].set_title('minutes_watched_2021_paid_1')

sb.kdeplot(df_3_filtered['minutes_watched'], ax=axes[1, 0])
axes[1, 0].set_title('minutes_watched_2022_paid_0')

sb.kdeplot(df_4_filtered['minutes_watched'], ax=axes[1, 1])
axes[1, 1].set_title('minutes_watched_2022_paid_1')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
# Saving each filtered DataFrame to a CSV file with specified naming format

df_1_filtered.to_csv("minutes_watched_2021_paid_0_no_outliers.csv", index=False)
df_2_filtered.to_csv("minutes_watched_2021_paid_1_no_outliers.csv", index=False)
df_3_filtered.to_csv("minutes_watched_2022_paid_0_no_outliers.csv", index=False)
df_4_filtered.to_csv("minutes_watched_2022_paid_1_no_outliers.csv", index=False)
```
